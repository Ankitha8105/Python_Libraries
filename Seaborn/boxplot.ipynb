### Q1. Write a program to draw box plot of life expectancy by continent for a data set


```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
def draw_boxplot():
    """ 
    Description :
        This function is to draw box plot of life expectancy by continent for a data set
    Return :
        It returns box plot of life expectancy by continent for a data set
    """
    dataset = pd.read_csv('FiveYearData.csv')
    sns.catplot(x='continent',y='lifeExp',data=dataset,kind='box')
    
    
def main():
    draw_boxplot()
    
    
if __name__ == "__main__":
    main()
```
