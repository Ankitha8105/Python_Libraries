### Q1. Write a program to draw bar plot of sex against survived for a dataset

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
def draw_barplot():
    """ 
    Description :
        This function is used to draw bar plot of sex against survived for a dataset
    Return :
        It returns bar plot of sex against survived for a dataset
    """
    dataset = pd.read_csv('titanic.csv')
    sns.catplot(x='sex',y='survived',data=dataset,hue='pclass',kind='bar')
    
    
def main():
    draw_barplot()
    
    
if __name__ == "__main__":
    main()
```
