### Q1. Write a program to draw a swarm plot of total bill against size for a given dataset

```python
import pandas as pd
import seaborn as sns
```

```python
def draw_swarm_plot():
    """ 
    Description :
        This function is to draw a swarm plot of total bill against size for a given dataset
    Return :
        It returns swarm plot of total bill against size for a given dataset
    """
    dataset = pd.read_csv('tips.csv')
    sns.catplot(x='size',y='total_bill',data=dataset,hue='sex',kind='swarm',s=15)
    
    
def main():
    draw_swarm_plot()
    
    
if __name__ == "__main__":
    main()
```
