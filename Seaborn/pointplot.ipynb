### Q1. Write a program to draw a point plot for sex against survived for a dataset

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
def draw_pointplot():
    dataset = pd.read_csv('titanic.csv')
    sns.catplot(x='sex',y='survived',data=dataset,hue='alone',kind='point')
    
    
def main():
    draw_pointplot()
    
    
if __name__ == "__main__":
    main()
```
