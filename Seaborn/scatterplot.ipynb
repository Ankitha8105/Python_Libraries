### Q1. Write a program to draw a scatter plot of “day” against “total bill” for a dataset

```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
def draw_scatter_plot():
    """ 
    Description :
        This function is to draw a scatter plot of “day” against “total bill” for a dataset
    Return :
        It returns scatter plot of “day” against “total bill” for a dataset
    """
    dataset = pd.read_csv('tips.csv')
    sns.relplot(x='day',y='total_bill',hue='size',data=dataset)
    
    
def main():
    draw_scatter_plot()
    
    
if __name__ =="__main__":
    main()
```
