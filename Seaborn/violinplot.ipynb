### Q1. Write a program to draw a violin plot of sex against total_bill for a given dataset


```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
def draw_violin_plot():
    """ 
    Description :
        This function is to draw a violin plot of sex against total_bill for a given dataset
    Return :
        It returns violin plot of sex against total_bill for a given dataset
    """
    dataset = pd.read_csv('tips.csv')
    sns.catplot(x='sex',y='total_bill',data=dataset,hue='time',kind='violin')
    
    
def main():
    draw_violin_plot()
    
    
if __name__ == "__main__":
    main()
```

### Q2. Write a program to draw a violin plot of “species” against “sepal length” for a dataset

```python
def generate_violin_plot():
    """ 
    Description :
        This function is to draw a violin plot of “species” against “sepal length” for a dataset
    Return :
        It returns violin plot of “species” against “sepal length” for a dataset
    """
    dataset = pd.read_csv('iris.csv')
    sns.catplot(x='species',y='sepal_length',data=dataset,kind='violin')


def main():
    generate_violin_plot()
    
    
if __name__ == "__main__":
    main()    
```
