### Q1. Write a Python programming to create a pie chart of the popularity of programming Languages

```python
import matplotlib.pyplot as plt
import logger
import numpy as np
import pandas as pd
```

```python
log = logger.logger_init('Q1')


def create_piechart(popularity):
    """ 
    Description :
        This function is used to create a pie chart of the popularity of programming Languages
    Parameters :
        popularity = List of popularity of each languages
    Return :
        It returns piechart
    """
    plt.figure(figsize=(7,5))
    plt.pie(popularity,explode=[0,0,0,0,0.2,0],shadow=True,autopct='%2.1f%%')

    plt.title("popularity of programming Languages",color='navy')
    plt.show()
    
    
def main():
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
    create_piechart(popularity)
    

if __name__ == "__main__":
    main()
```

### Q2. Write a Python programming to create a pie chart with a title of the popularity of programming Languages

```python
log = logger.logger_init('Q1')


def labeled_piechart(lang,popularity):
    """ 
    Description :
        This function is used to create a pie chart with a title of the popularity of programming Languages
    Parameters :
        lang = List of programming languages
        popularity = List of popularity of each languages
    Return :
        It returns Labeled piechart
    """
    plt.figure(figsize=(6,4))
    plt.pie(popularity,labels=lang,shadow=True)

    plt.title("popularity of programming Languages",color='navy')
    plt.show()
    
    
def main():
    lang = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
    labeled_piechart(lang,popularity)
    

if __name__ == "__main__":
    main()
```

### Q3. Write a Python programming to create a pie chart of gold medal achievements of five most successful countries in 2016 Summer Olympics. Read the data from a csv file

```python
log = logger.logger_init('Q3')


def read_from_csv():
    """ 
    Description :
        This function is used to create a pie chart of gold medal achievements of five most
        successful countries in 2016 Summer Olympics. Read the data from a csv file
    Return :
        It returns piechart
    """
    plt.figure(figsize=(8,5))
    data = pd.read_csv('medal.csv')
    plt.pie(data.gold_medal,labels=data.country,shadow=True)
    
    plt.title("Olympics Gold Medal achievements",color='purple')
    plt.legend(loc='upper left')
    plt.show()
    
    
def main():
    read_from_csv()
    
    
if __name__ == "__main__":
    main()
```
