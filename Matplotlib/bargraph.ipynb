### Q1. Write a Python programming to display a bar chart of the popularity of programming Languages

```python
import matplotlib.pyplot as plt
import logger
import numpy as np
import pandas as pd
```

```python
log = logger.logger_init('Q1')


def display_barchart(languages,popularity):
    """ 
    Description :
        This function is used to display a bar chart of the popularity of programming
        Languages
    Parameters :
        languages = List of programming languages
        popularity = List of popularity of each languages
    Return :
        It returns Bar graph
    """
    plt.figure(figsize=(6,4))
    plt.bar(languages,popularity,color='chocolate',label='popularity ',width=0.5)
    
    plt.xlabel("Programming languages")
    plt.ylabel("popularity")
    plt.title("popularity of programming")
    
    plt.legend()
    plt.show()
    
    
def main():
    lang = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
    display_barchart(lang,popularity)
    lang
    
if __name__ == "__main__":
    main()
```

### Q5. Write a Python programming to display a bar chart of the popularity of programming Languages. Attach a text label above each bar displaying its popularity  

```python
log = logger.logger_init('Q5')


def label_barchart(languages,popularity):
    """ 
    Description :
        This function is used to o display a bar chart of the popularity of programming
        Languages. Attach a text label above each bar displaying its popularity.
    Parameters :
        languages = List of programming languages
        popularity = List of popularity of each languages
    Return :
        It returns Bar graph of text label above each bar displaying its popularity
    """
    
    plt.figure(figsize=(7,5))
    plt.bar(languages,popularity,color='SkyBlue',label='popularity',width=0.5)
    
    for idx,val in enumerate(popularity):
        plt.text(idx,val+0.4,f'{val:.1f}', ha='center')
    plt.xlabel("Programming languages")
    plt.ylabel("popularity")
    plt.title("popularity of programming")
    
    plt.legend()
    plt.show()
    
    
def main():
    lang = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
    label_barchart(lang,popularity)
    
    
if __name__ == "__main__":
    main()
```

### Q6. Write a Python programming to display a bar chart of the popularity of programming Languages. Make blue border to each bar

```python
log = logger.logger_init('Q6')


def blue_edge_barchart(languages,popularity):
    """ 
    Description :
        This function is used to o display a bar chart of the popularity of programming
        Languages.Make blue border to each bar.
    Parameters :
        languages = List of programming languages
        popularity = List of popularity of each languages
    Return :
        It returns blue border to each bar
    """
    
    plt.figure(figsize=(7,5))
    plt.bar(languages,popularity,color='LimeGreen',edgecolor = 'blue',label='popularity',width=0.5)
    
    plt.xlabel("Programming languages")
    plt.ylabel("popularity")
    plt.title("popularity of programming")
    
    plt.minorticks_on()
    plt.grid(which='major',linestyle="solid",linewidth=0.4,color='Maroon')
    plt.grid(which='minor',linestyle="--",linewidth=0.4,color='LightCoral')
    
    plt.legend()
    plt.show()
    
    
def main():
    lang = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
    blue_edge_barchart(lang,popularity)
    
    
if __name__ == "__main__":
    main()
```

### Q7. Write a Python programming to display a bar chart of the popularity of programming Languages. Specify the position of each bar plot

```python
log = logger.logger_init('Q6')


def specified_pos_barplot(languages,popularity):
    """ 
    Description :
        This function is used to o display a bar chart of the popularity of programming
        Languages.Specify the position of each bar plot.
    Parameters :
        languages = List of programming languages
        popularity = List of popularity of each languages
    Return :
        It returns bar plots on Specified the position
    """
    y_positions = [1,3,4,5,7,9]
    plt.figure(figsize=(6,4))
    plt.bar(y_positions,popularity,color='DarkOrchid',label='popularity',width=0.5)
    
    plt.xlabel("Programming languages")
    plt.ylabel("popularity")
    plt.title("popularity of programming")
    plt.xticks(y_positions, languages)
     
    plt.legend()
    plt.show()
    
    
def main():
    lang = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
    specified_pos_barplot(lang,popularity)
    
    
if __name__ == "__main__":
    main()
```
