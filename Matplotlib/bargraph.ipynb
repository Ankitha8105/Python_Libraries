### Q1. Write a Python programming to display a bar chart of the popularity of programming Languages

```python
import matplotlib.pyplot as plt
import logger
import numpy as np
import pandas as pd
```

```python
log = logger.logger_init('Q1')


def display_barchart(languages,popularity):
    """ 
    Description :
        This function is used to display a bar chart of the popularity of programming
        Languages
    Parameters :
        languages = List of programming languages
        popularity = List of popularity of each languages
    Return :
        It returns Bar graph
    """
    plt.figure(figsize=(6,4))
    plt.bar(languages,popularity,color='chocolate',label='popularity ',width=0.5)
    
    plt.xlabel("Programming languages")
    plt.ylabel("popularity")
    plt.title("popularity of programming")
    
    plt.legend()
    plt.show()
    
    
def main():
    lang = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
    display_barchart(lang,popularity)
    lang
    
if __name__ == "__main__":
    main()
```

### Q5. Write a Python programming to display a bar chart of the popularity of programming Languages. Attach a text label above each bar displaying its popularity  

```python
log = logger.logger_init('Q5')


def label_barchart(languages,popularity):
    """ 
    Description :
        This function is used to o display a bar chart of the popularity of programming
        Languages. Attach a text label above each bar displaying its popularity.
    Parameters :
        languages = List of programming languages
        popularity = List of popularity of each languages
    Return :
        It returns Bar graph of text label above each bar displaying its popularity
    """
    
    plt.figure(figsize=(7,5))
    plt.bar(languages,popularity,color='SkyBlue',label='popularity',width=0.5)
    
    for idx,val in enumerate(popularity):
        plt.text(idx,val+0.4,f'{val:.1f}', ha='center')
    plt.xlabel("Programming languages")
    plt.ylabel("popularity")
    plt.title("popularity of programming")
    
    plt.legend()
    plt.show()
    
    
def main():
    lang = np.array(['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++'])
    popularity = np.array([22.2, 17.6, 8.8, 8, 7.7, 6.7])
    label_barchart(lang,popularity)
    
    
if __name__ == "__main__":
    main()
```

### Q9. Write a Python programming to display a bar chart of the popularity of programming Languages. Increase bottom margin.

### Q10. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women

```python
log = logger.logger_init('Q10')


def men_women_chart(men_score,women_score,n_grp):
    """ 
    Description :
        This function is used to create bar plot of scores by group and gender. Use multiple X
        values on the same chart for men and women.
    Parameters :
        men_score = tuple of men score
        women_score = tuple if women score
        n_grp = number of groups
    Return :
        It returns bar plots on the same chart for men and women.
    """
    plt.figure(figsize=(6,4))
    index = np.arange(n_grp)
    bar_width = 0.15
    plt.bar(index,men_score,color='red',alpha=0.8,label="Men",width=bar_width)
    plt.bar(index+bar_width,women_score,color="green",alpha=0.8,label="Women",width=bar_width)     
    
    plt.xlabel("Scores")
    plt.ylabel("Groups")
    plt.title("Score By Group")
    
    plt.xticks(index + bar_width, ('G1', 'G2', 'G3', 'G4', 'G5'))
    plt.legend()
    plt.show()
    
    
def main():
    n_groups = 5
    men_means = (22, 30, 33, 30, 26)
    women_means = (25, 32, 30, 35, 29)
    men_women_chart(men_means,women_means,n_groups)
    
    
if __name__ == "__main__":
    main()
```

### Q11. Write a Python program to create bar plot from a DataFrame

### Q12. Write a Python program to create bar plots with error bars on the same figure

```python
log = logger.logger_init('Q12')


def plot_with_error(velocity,errors):
    """ 
    Description :
        This function is used to create bar plots with error bars on the same figure
    Parameters :
        velocity = mean velocity
        errors = Standard deviation of velocity
    Return :
        It returns bar plots with error bars on the same figure
    """
    n=4
    ind = np.arange(n)  
    
    width = 0.25   
    plt.bar(ind, velocity, width, yerr=errors, color='red',label='plot with errors')
    
    plt.ylabel('Scores')
    plt.xlabel('Velocity')
    plt.title('Scores by Velocity')
    plt.legend()
    
    plt.show()
    
    
def main():
    mean_velocity = np.array([0.2474, 0.1235, 0.1737, 0.1824])
    std_error = np.array([0.3314, 0.2278, 0.2836, 0.2645])
    plot_with_error(mean_velocity,std_error)
    
    
if __name__ == "__main__":
    main()
```

### Q13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text label above each bar displaying men means
