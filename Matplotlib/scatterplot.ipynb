### Q1. Write a Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other

```python
import matplotlib.pyplot as plt
import logger
import numpy as np
import pandas as pd
```

```python
log = logger.logger_init('Q1')


def draw_scatter(x,y):
    """ 
    Description :
        This function is used to  draw a scatter graph taking a random distribution in X and Y and
        plotted against each other
    Parameters :
        x = First Random Distribution List
        y = Second Random Distribution List
    Return :
        It returns Scatterplots
    """
    plt.figure(figsize=(6,4))
    plt.scatter(x,y,color='navy',alpha=0.7,s=36)
    
    plt.xlabel('x-axis',fontsize=18)
    plt.ylabel('y-axis',fontsize=18)
    plt.title("Random Distribution Graph",fontsize=18,color='purple')
    plt.show()
    
    
def main():
    x = np.random.rand(25)
    y = np.random.rand(25)
    draw_scatter(x,y)
    

if __name__ == "__main__":
    main()
    
```

### Q2. Write a Python program to draw a scatter plot with empty circles taking a random distribution in X and Y and plotted against each other

```python
log = logger.logger_init('Q2')


def empty_circle_scatter(x,y):
    """ 
    Description :
        This function is used to draw a scatter plot with empty circles taking a random distribution
        in X and Y and plotted against each other
    Parameters :
        x = First Random Distribution List
        y = Second Random Distribution List
    Return :
        It returns Scatterplots
    """
    plt.figure(figsize=(5,3))
    plt.scatter(x,y,color='navy',alpha=0.7,s=56,marker="*",facecolors='none',edgecolors='g')
    
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.title("Random Distribution Graph",fontsize=15,color='purple')
    plt.show()
    
    
def main():
    x = np.random.randint(1,100,20)
    y = np.random.randint(10,199,20)
    empty_circle_scatter(x,y)
    

if __name__ == "__main__":
    main()
    
```
