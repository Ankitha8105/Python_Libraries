### Q1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import logger
```

```python
log = logger.logger_init('Q1')


def draw_line(x,y):
    """ 
    Description :
        This function is used to draw a line with suitable label in the x axis, y axis and a title
    Parameters :
        x = First Numpy array elements
        y = Second Numpy array
    Return :
        It returns graph
    """
    plt.figure(figsize=(5,3))
    line = plt.plot(x,y,label = "Straight Line")
    
    plt.title("Simple Example Graph")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    
    plt.xticks([2,4,6,8,10])
    plt.yticks([1,3,5,7,9])
    
    plt.legend()
    
    plt.show()
    
    
def main():
    x = np.array([1,3,5,7])
    y = np.array([2,4,6,8])
    draw_line(x,y)
    

if __name__ == "__main__":
    main()
```

### Q3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title

```python
log = logger.logger_init('Q3')


def plot_text_data():
    """ 
    Description :
        This function is used to draw a line using given axis values taken from a text file, with
        suitable label in the x axis, y axis and a title
    Return :
        It returns line of given values
    """
    plt.figure(figsize=(5,3))
    file_path = r'C:\Users\dell\Python_Libraries\Matplotlib\test.txt'
    
    with open(file_path ,'r') as file:
        data = file.readlines()
    
    x =[]
    y =[]
    for line in data:
        parts = line.split()
        x.append(int(parts[0]))  
        y.append(int(parts[1]))
        
    plt.plot(x,y,label="File Data") 
    plt.title("Plot of Data from test.txt")
    plt.xlabel("x values")
    plt.ylabel("y values")
    
    plt.legend()
    plt.show()
    

def main():
    plot_text_data()
    

if __name__ == "__main__":
    main()
    
```

### Q4. Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016

### Q5. Write a Python program to plot two or more lines on same plot with suitable legends of each line

```python
log = logger.logger_init('Q5')


def plot_twolines(x):
    """ 
    Description :
        This function is used to plot two or more lines on same plot with suitable legends of each
        line
    Parameter:
        x = List values
    Return :
        It returns line of given values
    """
    plt.figure(figsize=(6,4))
    w = [i*2 for i in x]
    y = [i*i for i in x]
    z = [i*4 for i in x]
    
    plt.plot(x,y,label="First Line",ls = ":",marker='o',ms=3)
    plt.plot(x,z,label = "Second line",ls="--",marker = 'o',ms=3)
    plt.plot(x,w,label="Third Line",ls = "solid",marker='o',ms=3)
      
    plt.title("Three Lines in one plot") 
    plt.xlabel("x-axis") 
    plt.ylabel("y-axis")
    
    plt.legend()
    plt.show()
    
    
def main():
    x = np.arange(1,6)
    plot_twolines(x)
    
    
if __name__ == "__main__":
    main()
    
```

### Q6. Write a Python program to plot two or more lines with legends, different widths and colors.

```python
log = logger.logger_init('Q6')


def lines_different_width(x):
    """ 
    Description :
        This function is used to plot two or more lines with legends, different widths and colors.
    Parameter:
        x = List values
    Return :
        It returns lines of given values
    """
    plt.figure(figsize=(6,4))
    w = [i*6 for i in x]
    y = [i*5 for i in x]
    z = [i*3 for i in x]
    
    plt.plot(x,y,label="First Line",color = 'purple',lw='3')
    plt.plot(x,z,label = "Second line",color ='hotpink',lw=4,)
    plt.plot(x,w,label="Third Line",color ='orange',lw =2)
      
    plt.title("Lines with different width and color") 
    plt.xlabel("x-axis") 
    plt.ylabel("y-axis")
    
    plt.legend()
    plt.show()
    
    
def main():
    x = np.arange(1,15,2)
    lines_different_width(x)
    
    
if __name__ == "__main__":
    main()
    
```

### Q7. Write a Python program to plot two or more lines with different styles

```python
log = logger.logger_init('Q7')


def lines_different_style(x):
    """ 
    Description :
        This function is used to plot two or more lines with different styles
    Parameter:
        x = List values
    Return :
        It returns lines of given values
    """
    plt.figure(figsize=(6,4))
    w = [i*6 for i in x]
    y = [i*i for i in w]
    z = [i*3 for i in y]
    
    plt.plot(x,y,label="First Line",color = 'blue',ls =":")
    plt.plot(x,z,label = "Second line",color ='hotpink',ls ="--")
    plt.plot(x,w,label="Third Line",color ='brown',ls="-.")
      
    plt.title("Lines with different styles") 
    plt.xlabel("x-axis") 
    plt.ylabel("y-axis")
    
    plt.legend()
    plt.show()
    
    
def main():
    x = np.arange(1,20,3)
    lines_different_style(x)
    
    
if __name__ == "__main__":
    main()
    
```

### Q8. Write a Python program to plot two or more lines and set the line markers

```python
log = logger.logger_init('Q8')


def lines_different_markers(u,v):
    """ 
    Description :
        This function is used to plot two or more lines and set the line markers
    Parameter:
        u = List values
        v = List values
    Return :
        It returns lines of given styles
    """
    plt.figure(figsize=(6,4))
    w = [i*6 for i in u]
    y = [i*i for i in v]
    z = [i*3 for i in w]
    
    plt.plot(u,w,label="First Line",color = 'brown',ls =":",marker="H",ms=3)
    plt.plot(y,v,label = "Second line",color ='hotpink',ls ="--",marker='d',ms=3)
    plt.plot(z,u,label="Third Line",color ='darkblue',ls="-.",marker = "*",ms=5)
    plt.plot(z,w,label="Fourth Line",color ='green',ls="-.",marker = "+",ms=5)
      
    plt.title("Lines with different markers") 
    plt.xlabel("x-axis") 
    plt.ylabel("y-axis")
    
    plt.legend()
    plt.show()
    
    
def main():
    x = np.arange(1,20,3)
    y = np.arange(0,12,2)
    lines_different_markers(x,y)
    
    
if __name__ == "__main__":
    main()
    
```

### Q9. Write a Python program to display the current axis limits values and set new axis values

```python
log = logger.logger_init('Q9')


def change_axis(x,y):
    """ 
    Description :
        This function is used to display the current axis limits values and set new axis values
    Parameter:
        x = First List values
        y = Second List values
    Return :
        It returns lines of given axis limits
    """
    plt.figure(figsize=(5,3))
    plt.plot(x,y,label="Sample data")
    axis = plt.axis()
    print(f"x axis limit is:{axis[:2]}")
    print(f"Y axis limit is:{axis[2:]}")
    
    plt.xlim(0,6)
    plt.ylim(0,60)
    
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Setting New Axis Limits')
    plt.legend()
    plt.show()
    
    
def main():
    x=np.array([1, 2, 3, 4, 5])
    y=np.array([10,20,30,40,50])
    change_axis(x,y)
    
    
if __name__ == "__main__":
    main()
```

### Q10. Write a Python program to plot quantities which have an x and y position

### Q11. Write a Python program to plot several lines with different format styles in one command using arrays

```python
log = logger.logger_init('Q11')


def plot_in_onecmd(x):
    """ 
    Description :
        This function is used to plot several lines with different format styles in one command
        using arrays
    Parameters :
        x = Numpy array elements
    Return :
        It returns graph of several lines with different format styles in one command using arrays
    """
    plt.figure(figsize=(5,3))
    plt.plot(x,x,'b:',x,x**2,'r--',x,x**3,'g-.')
    
    plt.title("plot 3 lines in single command")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    
    plt.show()
    
def main():
    x = np.arange(0,2,0.2)
    plot_in_onecmd(x)
    
    
if __name__=="__main__":
    main()
```

### Q12. Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with linestyle -, width .5. and color blue

### Q13. Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn off ticks

```python
log = logger.logger_init('Q4')
from matplotlib.dates import date2num
import datetime as DT

def display_grids():
    """ 
    Description :
        This function is used to display the grid and draw line charts of the closing value of
        Alphabet Inc. between October 3, 2016 to October 7, 2016. 
    Return :
        It returns line of given values
    """

    data = pd.read_csv('fdata.csv')
    fig = plt.figure(figsize=(5,3))

    plt.plot(data.Date,data.Close,'r-o')
    plt.xlabel('Date')
    plt.ylabel('Closing Value') 
    plt.title('Closing stock value') 
   
    plt.minorticks_on()
    plt.grid(which='major', linestyle='-', linewidth='0.5', color='orange')
    plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

    plt.show()
       
    
def main():
    display_grids()
    
    
if __name__ == "__main__":
    main()
```
