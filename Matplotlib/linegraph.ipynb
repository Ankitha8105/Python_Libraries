### Q1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import logger
```

```python
log = logger.logger_init('Q1')


def draw_line(x,y):
    """ 
    Description :
        This function is used to draw a line with suitable label in the x axis, y axis and a title
    Parameters :
        x = First Numpy array elements
        y = Second Numpy array
    Return :
        It returns graph
    """
    plt.figure(figsize=(5,3))
    line = plt.plot(x,y,label = "Straight Line")
    
    plt.title("Simple Example Graph")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    
    plt.xticks([2,4,6,8,10])
    plt.yticks([1,3,5,7,9])
    
    plt.legend()
    
    plt.show()
    
    
def main():
    x = np.array([1,3,5,7])
    y = np.array([2,4,6,8])
    draw_line(x,y)
    

if __name__ == "__main__":
    main()
```

### Q2. Write a Python program to draw a line using given axis values with suitable label in the x axis, y axis and a title

```python
log = logger.logger_init('Q2')


def values_line(x,y):
    """ 
    Description :
        This function is used to draw a line using given axis values with suitable label in the x axis, y axis and a title
    Parameters :
        x = First Numpy array elements
        y = Second Numpy array
    Return :
        It returns graph
    """
    plt.figure(figsize=(5,3))
    line = plt.plot(x,y,label = "curve")
    
    plt.title("Simple Graph")
    plt.xlabel("x-axis")
    plt.ylabel("y-axis")
    
    plt.xticks([2,4,6,8,10])
    z = np.arange(0,100,10)
    plt.yticks(z)
    
    plt.legend()
    
    plt.show()
    
    
def main():
    x = np.arange(1,10)
    y = [i*i for i in x]
    
    values_line(x,y)
    

if __name__ == "__main__":
    main()
```

### Q3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title

```python
log = logger.logger_init('Q3')


def plot_text_data():
    """ 
    Description :
        This function is used to draw a line using given axis values taken from a text file, with
        suitable label in the x axis, y axis and a title
    Return :
        It returns line of given values
    """
    plt.figure(figsize=(5,3))
    file_path = r'C:\Users\dell\Python_Libraries\Matplotlib\test.txt'
    
    with open(file_path ,'r') as file:
        data = file.readlines()
    
    x =[]
    y =[]
    for line in data:
        parts = line.split()
        x.append(int(parts[0]))  
        y.append(int(parts[1]))
        
    plt.plot(x,y,label="File Data") 
    plt.title("Plot of Data from test.txt")
    plt.xlabel("x values")
    plt.ylabel("y values")
    
    plt.legend()
    plt.show()
    

def main():
    plot_text_data()
    

if __name__ == "__main__":
    main()
    
```

### Q4. Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016

```python
log = logger.logger_init('Q4')


def plot_csv_data():
    """ 
    Description :
        This function is used to draw line charts of the financial data of Alphabet Inc. between
        October 3, 2016 to October 7, 2016
    Return :
        It returns line of given values
    """
    plt.figure(figsize=(6,4))
    data = pd.read_csv('fdata.csv')
    
    plt.plot(data.Date,data.Open,label = "Open",ls = "dotted",marker = "*",ms = 5)
    plt.plot(data.Date,data.High,label = "High",ls = "dashed",marker = 'o',ms = 3)
    plt.plot(data.Date,data.Low,label = "Low",marker= '*',ms = 5)
    plt.plot(data.Date,data.Close,label = "Close",ls = "-.",marker = "o",ms = 3)
    
    plt.title("Sample Financial data")
    plt.xlabel("Dates")
    plt.ylabel("Price")
    plt.legend()            
    plt.show()
    
    
def main():
    plot_csv_data()
    
    
if __name__ == "__main__":
    main()
```

### Q5. Write a Python program to plot two or more lines on same plot with suitable legends of each line

```python
log = logger.logger_init('Q5')


def plot_twolines(x):
    """ 
    Description :
        This function is used to plot two or more lines on same plot with suitable legends of each
        line
    Parameter:
        x = List values
    Return :
        It returns line of given values
    """
    w = [i*2 for i in x]
    y = [i*i for i in x]
    z = [i*4 for i in x]
    
    plt.plot(x,y,label="First Line",ls = ":",marker='o',ms=3)
    plt.plot(x,z,label = "Second line",ls="--",marker = 'o',ms=3)
    plt.plot(x,w,label="Third Line",ls = "solid",marker='o',ms=3)
      
    plt.title("Two Lines in one plot") 
    plt.xlabel("x-axis") 
    plt.ylabel("y-axis")
    
    plt.legend()
    plt.show()
    
    
def main():
    x = np.arange(1,6)
    plot_twolines(x)
    
    
if __name__ == "__main__":
    main()
    
```
