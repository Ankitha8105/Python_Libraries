### Q1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
import plotly.express as px
import numpy as np
import pandas as pd
```

```python
def draw_scatter(df):
    """ 
    Description :
        This function is used to draw a scatter plot for random 1000 x and y coordinates
    Parameters :
        df = DataSet
    Return :
        It returns scatter graph
    """
    fig = px.scatter(df,x='x',y='y',title='Scatter Plot of Random Coordinates')
    fig.show()
    
    
def main():
    x = np.random.randint(1,100,1000)
    y = np.random.randint(100,900,1000)
    dic = pd.DataFrame({'x':x,'y':y})
    draw_scatter(dic)
    
    
if __name__ == "__main__":
    main()
```

### Q2. Write a program to draw line and scatter plots for random 100 x and y coordinates


```python
def line_with_scatterplot(df):
    """ 
    Description :
        This function is used to draw line and scatter plots for random 100 x and y coordinates
    Parameters :
        df = DataSet
    Return :
        It returns scatter graph of line and scatter plots for random 100 x and y coordinates
    """
    line_fig = px.line(df, x='x', y='y', markers=True)
    line_fig.show()
    
    
def main():
    x = np.random.rand(100)
    y = np.random.rand(100)
    dic = pd.DataFrame({'x':x,'y':y})
    line_with_scatterplot(dic)
    
    
if __name__ == "__main__":
    main()
```

### Q3. Write a program to draw a scatter plot for random 500 x and y coordinates and style it


```python
def generate_scatter(df):
    """ 
    Description :
        This function is used to draw a scatter plot for random 500 x and y coordinates and style it
    Parameters :
        df = DataSet
    Return :
        It returns scatter graph for random 500 x and y coordinates
    """
    fig = px.scatter(df,x='x',y='y',title='Styled Scatter Plot of Random Coordinates',
        labels={'x': 'X Coordinate', 'y': 'Y Coordinate'},
        color_discrete_sequence=['maroon'],  
        )
    fig.show()
    
    
def main():
    x = np.random.randint(500,950,1000)
    y = np.random.randint(200,800,1000)
    dic = pd.DataFrame({'x':x,'y':y})
    generate_scatter(dic)
    
    
if __name__ == "__main__":
    main()
    
```

### Q4. Write a program to draw a scatter plot for a given dataset and show datalabels on hover

```python
def show_datalables(df):
    """ 
    Description :
        This function is used to draw a scatter plot for a given dataset and show datalabels on hover
    Parameters :
        df = DataSet
    Return :
        It returns scatter graph for a given dataset and show datalabels on hover
    """
    fig = px.scatter(
        df, 
        x='Postal', 
        y='State', 
        title='Scatter Plot with Hover Labels for USA States',
        hover_data={'Postal': True, 'State': True, 'Population': True}  
    )
    fig.show()
    
    
def main():
    data = pd.read_csv('usa_states.csv')
    show_datalables(data)
    
    
if __name__ == "__main__":
    main()
    
    
```
